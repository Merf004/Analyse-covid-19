#### Importation des bibliothèques

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Importation du jeu de données

```python
data = pd.read_csv('Covid Data.csv')
```

```python
data
```

```python
data.info()
```

```python
data.describe()
```

#### Pretraitement des données

```python
data = data.replace({97: np.nan, 99: np.nan})
```

```python
data = data.replace({98: 'NON'})
```

```python
data
```

```python
data.info()
```

```python
data['SEX'].value_counts()
```

```python
data.columns
```

```python
data['SEX'] = data['SEX'].replace({1: 'FEMME', 2: 'HOMME'})
```

```python
data
```

```python
data['CLASIFFICATION_FINAL'][data['CLASIFFICATION_FINAL'] < 4] = 'POSITIVE'
```

```python
data
```

```python
data['CLASIFFICATION_FINAL'][data['CLASIFFICATION_FINAL'] != 'POSITIVE'] = 'NEGATIVE'
```

```python
data
```

```python
data['PATIENT_TYPE'].replace({1:'DOMICILE', 2: 'HOSPITALISATION'}, inplace=True)
```

```python
data
```

```python
data.columns
```

```python
data[['PNEUMONIA', 'PREGNANT', 'DIABETES', 'COPD', 'ASTHMA', 'INMSUPR',
       'HIPERTENSION', 'OTHER_DISEASE', 'CARDIOVASCULAR', 'OBESITY',
       'RENAL_CHRONIC', 'TOBACCO']] = data[['PNEUMONIA', 'PREGNANT','DIABETES', 'COPD', 'ASTHMA', 'INMSUPR',
       'HIPERTENSION', 'OTHER_DISEASE', 'CARDIOVASCULAR', 'OBESITY',
       'RENAL_CHRONIC', 'TOBACCO']].replace({1:'OUI', 2:'NON'})
```

```python
data
```

```python
data['DATE_DIED'].replace({'9999-99-99': 'NON'}, inplace=True)
```

```python
data
```

```python
data['DATE_DIED'][data['DATE_DIED'] != 'NON'] = 'OUI'
```

```python
data
```

```python
data.info()
```

```python
data = data.drop(['INTUBED', 'MEDICAL_UNIT', 'ICU', 'USMER'], axis = 1)
```

```python
data
```

```python
data['PNEUMONIA'].value_counts()
```

```python
data = data.dropna(subset=['PNEUMONIA'])
```

```python
data.info()
```

```python
data = data.dropna(subset=['AGE'])
```

```python
data.describe(include='object')
```

```python
data
```

#### Objectifs
- Décès par age et par sexe
- Personnes positives par age et par sexe
- Décès en fonction des symptomes et d'autres maladies
- Personnes positives en fonction des symptomes et d'autres maladies

#### Décès par age et par sexe

```python
data_died = data[data['DATE_DIED'] == 'OUI']
```

```python
data_died
```

```python
data_died_1 = data_died.groupby('SEX')['DATE_DIED'].count()
```

```python
data_died_1
```

```python
plt.bar(data_died_1.index, data_died_1.values, color='lightgreen')
plt.title('Nombre de décès par sexe')
plt.xlabel('Sexe')
plt.ylabel('Décès')

plt.show()
```

```python
data['AGE'] = data['AGE'].replace({'NON': 98})
```

```python
data_died_2 = data_died.groupby('AGE')['DATE_DIED'].count()
```

```python
plt.bar(data_died_2.index, data_died_2.values, color='lightgreen')

plt.title('Nombre de décès par age (Histogramme)')
plt.xlabel('Age')
plt.ylabel('Décès')

plt.show()
```

#### Personnes positives en fonction de l'age et du sexe

```python
data_pos = data[data['CLASIFFICATION_FINAL'] == 'POSITIVE']
```

```python
data_pos
```

```python
data_pos1 = data_pos.groupby('SEX')['CLASIFFICATION_FINAL'].count()
```

```python
plt.bar(data_pos1.index, data_pos1.values, color='lightgreen')
plt.title('Nombre de cas positifs par sexe')
plt.xlabel('Sexe')
plt.ylabel('Cas positifs')

plt.show()
```

```python
data_pos2 = data_pos.groupby('AGE')['CLASIFFICATION_FINAL'].count()

plt.bar(data_pos2.index, data_pos2.values, color='lightgreen')
plt.title('Nombre de cas positifs par age')
plt.xlabel('Age')
plt.ylabel('Cas positifs')

plt.show()
```

#### Décès en fonction des symptomes et des maladies

```python
data.columns
```

```python
maladies = ['PNEUMONIA', 'DIABETES', 'COPD', 'ASTHMA', 'INMSUPR', 'HIPERTENSION', 'OTHER_DISEASE', 'CARDIOVASCULAR', 'OBESITY', 'RENAL_CHRONIC', 'TOBACCO']
```

```python
i = 1
plt.figure(figsize=(16, 13))
for maladie in maladies:
    plt.subplot(3, 4, i)
    data_m1 = data_died.groupby(maladie)['DATE_DIED'].count()
    plt.bar(data_m1.index, data_m1.values, color='lightgreen')
    plt.xlabel(maladie)
    plt.ylabel('Nombre')
    i = i+1
```

```python
i = 1
plt.figure(figsize=(16, 13))
for maladie in maladies:
    plt.subplot(3, 4, i)
    data_m2 = data_pos.groupby(maladie)['CLASIFFICATION_FINAL'].count()
    plt.bar(data_m2.index, data_m2.values, color='lightgreen')
    plt.xlabel(maladie)
    plt.ylabel('Nombre')
    i = i+1
```

#### Décès et cas positifs en fonction du type de patient

```python
data_tp1 = data_pos.groupby('PATIENT_TYPE')['CLASIFFICATION_FINAL'].count()

plt.bar(data_tp1.index, data_tp1.values, color='lightgreen')
plt.title('Type de prise en charge pour les personnes positives')
plt.xlabel('TYpe de prise en charge')
plt.ylabel('Nombres de cas pris en charge')
```

```python
data_tp2 = data_died.groupby('PATIENT_TYPE')['CLASIFFICATION_FINAL'].count()

plt.bar(data_tp2.index, data_tp2.values, color='lightgreen')
plt.title('Type de prise en charge pour les personnes décédées')
plt.xlabel('TYpe de prise en charge')
plt.ylabel('Nombres de cas de décès')
```
